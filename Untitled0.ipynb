```python
import random

#Si el numero que sale con randmo es 1, la prediccion es que el numero sera impar
#y si es 2, la prediccion es que el numero será par.
prediccionMaquina = random.randint(1,3);
numeroIngresado = int(input("Ingrese un numero: "))

if(prediccionMaquina == 1):
  print("La prediccion de la maquina es impar")
else:
  print("La prediccion de la maquina es par")

if(numeroIngresado % 2 == 0):
  if(prediccionMaquina == 2):
    print("La maquina acerto que era par")
  else:
    print("La maquina se equivocó, el numero era par")

else:
    if(prediccionMaquina == 1):
      print("La maquina acerto que era impar")
    else:
      print("La maquina se equivocó, era impar")
```

```python
def cifrado_cesar(texto, desplazamiento):
    resultado = ""
    for i in range(len(texto)):
        char = texto[i]

        # Cifrar mayúsculas
        if char.isupper():
            resultado += chr((ord(char) + desplazamiento - 65) % 26 + 65)
        # Cifrar minúsculas
        elif char.islower():
            resultado += chr((ord(char) + desplazamiento - 97) % 26 + 97)
        # Mantener espacios y otros caracteres
        else:
            resultado += char
    return resultado

# Uso del código
mensaje = input("Ingrese el mensaje a cifrar: ")
llave = 13
cifrado = cifrado_cesar(mensaje, llave)
print(f"Cifrado: {cifrado}") # Salida: Krno Pxqgr
print(f"Descifrado: {cifrado_cesar(cifrado, -llave)}")

```

```python
import pandas as pd

ruta_archivo = "/content/compras_semana.csv"
datos = pd.read_csv(ruta_archivo)
datos ['total'] = datos['precio'] * datos['cantidad']


total_compras = datos['total'].sum()
print(f"El total de las compras es: ${total_compras}")


promedio_compras = datos['total'].mean()
print(f"El promedio de las compras es: ${promedio_compras}")

mediana_compras = datos['total'].median()
print(f"Mediana de la compra: ${mediana_compras}")

compra_minima = datos['total'].min()
print(f"La compra minima es: ${compra_minima}")

compra_maxima = datos['total'].max()
print(f"La compra maxima es: ${compra_maxima}")

std = datos['total'].std()
print(f"La desviacion estandar es: ${std}")

var = datos['total'].var()
print(f"La varianza es: ${var}")

count = datos['total'].count()
print(f"Registro conteo es: ${count}")

datos.describe()
```
